```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Provided data 
hours_studied = [10, 9, 3, 2, 12, 7, 5, 11, 8, 6, 4, 15, 10, 13, 2, 14, 6, 9, 7, 5]
exam_score = [95, 90, 75, 60, 98, 85, 70, 96, 88, 80, 65, 100, 95, 98, 55, 99, 80, 90, 75, 70]

# creating the scatter plot
plt.scatter(hours_studied, exam_score)

# Adding labels
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scored')
#Adding the tittle 
plt.title('Relationship Between Hours Studied and Exam Scores ')

# Show plot
plt.grid(True)
plt.show()
```
