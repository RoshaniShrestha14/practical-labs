=> Explanatroy Data Analysis is a critical initial phase in data scince project which involves the process of visually and statistically exploring a dataset to understand its characteristics, patterns, trends and potential issues that occurs. EDA has many roles in shaping the understanding and expectations for data sceince projects and some of the roles are listed below:
1. Data Collection and Cleaning: EDA starts with collecting dataset and cleaning the impurities present in the data which ensure data integrity. This involves in handling missing values, dealing with outliers and converting data into the suitable format for data analysis.
2. Feature Engineering: EDA gives insights that help with feature engineering, which is the process of developing new variables or modifying existing ones to imprpve predictions models. Data scientists can use visualization, correlation analyisis and statistical tests to find important factors, assess their correlations, and uncover potential interactions taht could improve model performance.
3. Descriptive Statistics: EDA starts with calculating the summary satistics such as mean, median, mode, standard deviation, minimum and maximum, etc. These statistics is important because it provide an overview of the central tendency and spread the data.
4. Data Visualization:Visualization of the data is done using charts, graphs, and plot which helps in identifying patterns, trends and relationships. Common visualizations include histograms, boxplots, scatterplots, heatmaps and bar chats etc.
5. Identifying patterns and Relationships: EDA uncovers patterns, trends and realationship within the data. Visualization such as histograms, scatterplot and heatmaps allows the data scientists to explore correlations, dependencies and clusters. These techniques helps in understanding the underlying structure of the data.
