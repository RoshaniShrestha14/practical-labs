```python
# Concatenate all DataFrames
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
# Handle missing values
if 'column_name' in merged_df.columns:
    merged_df['column_name'].fillna(merged_df['column_name'].mean(), inplace=True)
 # Fill missing categorical values with the mode
for column in merged_df.select_dtypes(include=['object']).columns:
    merged_df[column].fillna(merged_df[column].mode()[0], inplace=True)
```

```python
merged_df.head()
```

```python

# Convert 'Order Date' to datetime if it's not already
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], errors='coerce')


```

```python
# Drop rows with invalid 'Order Date'
merged_df.dropna(subset=['Order Date'], inplace=True)

```

```python
# Extract additional information from 'Order Date'
merged_df['Month'] = merged_df['Order Date'].dt.month
merged_df['Day of Week'] = merged_df['Order Date'].dt.dayofweek
merged_df['Hour'] = merged_df['Order Date'].dt.hour
```

```python
merged_df.head()
```

```python

# Convert 'Quantity_Ordered' and 'Price_Each' to numeric if they are not already
merged_df['Quantity Ordered'] = pd.to_numeric(merged_df['Quantity Ordered'], errors='coerce')
merged_df['Price Each'] = pd.to_numeric(merged_df['Price Each'], errors='coerce')

# Drop rows with invalid 'Quantity_Ordered' or 'Price_Each'
merged_df.dropna(subset=['Quantity Ordered', 'Price Each'], inplace=True)

# Create a 'Sales' column
merged_df['Sales'] = merged_df['Quantity Ordered'] * merged_df['Price Each']
```

```python
# Define the months for each season
seasonal_months = {
    'Winter': [12, 1, 2],
    'Spring': [3, 4, 5],
    'Summer': [6, 7, 8],
    'Fall': [9, 10, 11]
}

# Calculate total sales for each product in each season
seasonal_sales = {}
for season, months in seasonal_months.items():
    seasonal_sales[season] = merged_df[merged_df['Month'].isin(months)].groupby('Product')['Sales'].sum()

# Identify the top product for each season
top_products_by_season = {}
for season, sales in seasonal_sales.items():
    top_product = sales.idxmax()
    top_sales = sales.max()
    top_products_by_season[season] = (top_product, top_sales)

# Display the results
for season, (product, sales) in top_products_by_season.items():
    print(f"The top product in {season} is '{product}' with sales of ${sales:.2f}.")
# Plot the results
season_names = list(top_products_by_season.keys())
top_products = [top_products_by_season[season][0] for season in season_names]
top_sales = [top_products_by_season[season][1] for season in season_names]

plt.figure(figsize=(10, 6))
sns.barplot(x=season_names, y=top_sales, palette='viridis')
for i, product in enumerate(top_products):
    plt.text(i, top_sales[i], product, ha='center', va='bottom', color='black')
plt.title('Top Product Sales by Season in 2019')
plt.xlabel('Season')
plt.ylabel('Sales in USD')
plt.show()
```

```python
#EDA
# Monthly Sales
monthly_sales = merged_df.groupby('Month').sum()['Sales']
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='navy')
plt.title('Monthly Sales in 2019')
plt.xlabel('Month')
plt.ylabel('Sales in USD')
plt.show()
```

```python
# Sales by Day of Week
weekday_sales = merged_df.groupby('Day of Week').sum()['Sales']
plt.figure(figsize=(10, 6))
weekday_sales.plot(kind='bar', color='aqua')
plt.title('Sales by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Sales in USD')
plt.show()
```

```python
# Sales by Hour
hourly_sales = merged_df.groupby('Hour').sum()['Sales']
plt.figure(figsize=(10, 6))
hourly_sales.plot(kind='line')
plt.title('Sales by Hour')
plt.xlabel('Hour')
plt.ylabel('Sales in USD')
plt.grid()
plt.show()
```

```python
# Show the top 10 products by sales
top_products = merged_df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y=top_products.index, color='cyan')
plt.title('Top 10 Products by Sales')
plt.xlabel('Sales in USD')
plt.ylabel('Product')
plt.show()
```

```python
# Most common products sold
product_sales = merged_df.groupby('Product').sum()['Quantity Ordered']
plt.figure(figsize=(10, 6))
product_sales.plot(kind='bar', color='orange')
plt.title('Most Common Products Sold')
plt.xlabel('Product')
plt.ylabel('Quantity_Ordered')
plt.show()
```

```python
# Average sales per order by product
product_avg_sales = merged_df.groupby('Product').mean()['Sales']
plt.figure(figsize=(10, 6))
product_avg_sales.plot(kind='bar')
plt.title('Average Sales per Order by Product')
plt.xlabel('Product')
plt.ylabel('Average Sales in USD')
plt.show()
```

```python
# Quantity ordered per product
quantity_per_product = merged_df.groupby('Product')['Quantity Ordered'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=quantity_per_product.values, y=quantity_per_product.index, color='teal')
plt.title('Total Quantity Ordered per Product')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product')
plt.show()
```
