```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#dataset seaborn mah data set load garni function huncha
iris = sns.load_dataset('iris')
iris.head()
```

```python
#univariant bhanya euta matra variable liyera analysis garni=boxplot, histogram
#multivariant bhanya chain multiple variable lini 
# HISTOGRAM 
#to change color and size
plt.figure(figsize=(4,4))
plt.subplot(1, 2, 1)
plt.hist(iris['sepal_length'], color= "green")
#plt.subplot(#row, column , position)
plt.subplot(1, 2, 2)
plt.hist(iris['sepal_width'], color= "blue")
plt.show()
#plt.hist(iris.sepal_length)

```

```python
#BOXPLOT
plt.boxplot([iris.sepal_length, iris.petal_length], vert= False)

plt.show()
```

```python
#LINE GRAPH
#by variant analysis- euta bhanda badi ko analysis garniu
#plt.plot(iris.sepal_length, iris.sepal_width)
plt.figure(figsize=4,4)
a =[1, 2, 3 ,4 ,5 ,6]
b=[7, 8, 9,  10, 11, 12]
```

```python
#Scatterplot
plt.scatter(data=iris, x= "sepal_length", y ="sepal_width")
```

```python
#BAR PLOR for categorial and numerical 
plt.figure(figsize=(6,6))
plt.title("Bar graph species v/s petal_length")
plt. bar(iris["species"], iris["petal_length"], color="pink")
plt.xlabel("petal_length")
plt.ylabel("species")
plt.show()
#horizontal chahiyo bhaneuy plt.barh
```

```python
#PIE CHART
#plt.pie(iris.sepal_length, labels= iris.species)
#pivot table
pivot_table= pd.pivot_table(iris[["sepal_length", "species"]], index="species",aggfunc="sum")
plt.pie(pivot_table["sepal_length"], labels= pivot_table.index, explode=(0,0,0.05), autopct="%0.1f%%")  
plt.show()
```

```python
#How to plot using seaborn
plt.figure(figsize=(6,6))
# sns.heatmap(iris[["sepal_length","sepal_width"]])
sns.histplot(iris["sepal_length"])
plt.show()
```
